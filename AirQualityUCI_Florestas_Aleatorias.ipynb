```python
import pandas as pd

df = pd.read_excel('AirQualityUCI.xlsx')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X_generico = df.drop(['Date', 'Time'], axis=1).reset_index(drop=True)
```

```python
X_generico.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,8))
sns.heatmap(X_generico.corr(), annot=True)
plt.show()
```

```python
X_generico.columns
```

```python
X = X_generico.drop(['PT08.S2(NMHC)'], axis=1).reset_index(drop=True)
y = X_generico['PT08.S2(NMHC)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=123)
```

```python
print('Tamanho de X_train: ', X_train.shape, '\n')
print('Tamanho de X_test: ', X_test.shape, '\n')
print('Tamanho de y_train: ', y_train.shape, '\n')
print('Tamanho de y_test: ', y_test.shape, '\n')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

lista = list(range(300,501,10))

for i in lista:
  clf = RandomForestRegressor(n_estimators=i,
                             max_depth=3,
                             random_state=123,
                             n_jobs=-1)
  
  clf.fit(X_train, y_train)
  y_pred_train = clf.predict(X_train)
  y_pred_test = clf.predict(X_test)
  r_quadrado_train = r2_score(y_train, y_pred_train)
  r_quadrado_test = r2_score(y_test, y_pred_test)

  print(f'O R2 de treino é {r_quadrado_train} e o de teste é {r_quadrado_test}')
```
